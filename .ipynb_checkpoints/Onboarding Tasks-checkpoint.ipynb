```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imdb_top_1000.csv')
df.describe()
```

```python
df.head()
```

# What are the column names in the dataset?

```python
column_names = list(df.columns)
for column in column_names:
    print(column)
```

# What is the average rating of the movies?

```python
average_rating = df['IMDB_Rating'].mean()
print(average_rating)
```

# How many unique genres are there?


```python
unique_genres_count = df['Genre'].nunique()
print(unique_genres_count)
```

# Save your results as a csv file.

```python

results = pd.DataFrame({'Question': ['Column Names', 'Average Rating', 'Number of Unique Genres'],
                       'Answer': [column_names, average_rating, unique_genres_count]})
results.to_csv('results.csv', index=False)
rs = pd.read_csv('results.csv')
rs.head()
```

## Plot a histogram of the genre distribution of the movies. How many movies from each genre are there in this list.

```python

unique_genres = df['Genre'].unique()[:50]


count = [df['Genre'].value_counts()[genre] for genre in unique_genres ]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(unique_genres,count)
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Genre Distribution of Movies')
plt.xticks(rotation=90)
plt.show()
```

## Plot a scatterplot in this dataset for the number of movies vs year (Each year, how many movies made to this top 1000 list?)

```python
movies = df["Released_Year"].value_counts().nlargest(15)
plt.scatter(movies.index,movies.values)
plt.xlabel("year")
plt.ylabel("movies")
plt.title('movies vs year')
plt.xticks(rotation = 45)
plt.show()
```
